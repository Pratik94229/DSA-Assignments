```python
def twoSum(nums, target):
  sum=0
  left=0
  right=len(nums)-1
  nums.sort()
  while(left<right):
    sum=nums[left]+nums[right]
    if(sum==target):
      return left,right
    elif(sum>target):
      right-=1
    elif(sum<target):
      left+=1

  return []

```

```python
nums = [2, 7, 11, 15]
target = 9
result = twoSum(nums, target)
print(result)  # Expected Output: [0, 1]
```

```python
nums = [-3,1,2, 7,9, 11, 15]
target = 9
result = twoSum(nums, target)
print(result)  # Expected Output: [2, 3]
```

```python
def removeElement(nums, val):
    k = 0
    for i in range(len(nums)):
        if nums[i] != val:
            nums[k] = nums[i]
            k += 1
    return k
```

```python
nums = [3, 2, 2, 3]
val = 3
new_length = removeElement(nums, val)
print("New Length:", new_length)  # Output: New Length: 2
print("Modified nums:", nums[:new_length])  # Output: Modified nums: [2, 2]
```

```python
def searchInsert(nums, target):
    left, right = 0, len(nums) - 1
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return left
```

```python
nums = [1, 3, 5, 6]
target = 5
index = searchInsert(nums, target)
print(index)  # Expected Output: 2
```

```python
nums = [1, 3, 5, 6]
target = 2
index = searchInsert(nums, target)
print(index)  #Expected Output: 2
```

```python
def plusOne(digits):
    carry = 1
    for i in range(len(digits) - 1, -1, -1):
        digits[i] += carry
        if digits[i] == 10:
            digits[i] = 0
            carry = 1
        else:
            carry = 0
            break
    if carry == 1:
        digits.insert(0, carry)
    return digits
```

```python
digits = [1, 2, 9]
result = plusOne(digits)
print(result)  #Expected Output: [1, 2, 9]
```

```python
def merge_sorted_arrays(nums1, m, nums2, n):

    i = m - 1
    j = n - 1
    k = m + n - 1

    while i >= 0 and j >= 0:
        if nums1[i] > nums2[j]:
            nums1[k] = nums1[i]
            i -= 1
        else:
            nums1[k] = nums2[j]
            j -= 1
        k -= 1

    # Copying the remaining elements of nums2 if any.
    while j >= 0:
        nums1[k] = nums2[j]
        j -= 1
        k -= 1

```

```python
nums1 = [1, 2, 3, 0, 0, 0]
nums2 = [2, 5, 6]
m,n = 3,3

merge_sorted_arrays(nums1, m, nums2, n)
print(nums1) #Expected output [1, 2, 2, 3, 5, 6]
```

```python
def if_duplicate(nums):
  n=0
  m=len(nums)-1
  flag=False
  while(n<m):
    if(nums[n]==nums[m]):
      flag=True
      return flag
    else:
      n+=1  
  return flag    
```

```python
nums=[1,2,3,1]
print(if_duplicate(nums))
```

```python
nums=[1,2,3,5,6]
print(if_duplicate(nums))
```

```python
def moveZeroes(nums):
    p1 = 0
    p2 = 0
    
    while p2 < len(nums):
        if nums[p2] != 0:
            nums[p1], nums[p2] = nums[p2], nums[p1]
            p1 += 1
        p2 += 1
    
    nums[p1:] = [0] * (len(nums) - p1)
```

```python
nums = [0, 1, 0, 3, 12]

moveZeroes(nums)
print(nums)  # Expected Output: [1, 3, 12, 0, 0]
```

```python
def findErrorNums(nums):
    seen = set()
    duplicate = -1
    missing = -1
    
    for num in nums:
        if num not in seen:
            seen.add(num)
        else:
            duplicate = num
    
    for i in range(1, len(nums) + 1):
        if i not in seen:
            missing = i
    
    return [duplicate, missing]
```

```python
nums = [1, 2, 2, 4]

result = findErrorNums(nums)
print(result)  # Expected Output: [2, 3]
```
